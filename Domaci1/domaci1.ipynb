```python
def uHom(p):
    x = p[0]
    y = p[1]
    return np.array([float(x), float(y), 1.0])
```

```python
P1 = (829, 410)
P2 = (551, 659)
P3 = (286, 505)
P5 = (847, 271)
P6 = (554, 496)
P7 = (271, 355)
P8 = (593, 190)
```

```python
def prava(X, Y):
    return np.cross(X, Y)
```

```python
def presekPravih(p, q):
    A = np.cross(p, q)
    return A / p[2]
```

```python
def nadjiNepoznatu(tacke):
    p1, p2, p3, p5, p6, p7, p8 = [uHom(t) for t in tacke]

    # beskonacno daleka tacka po X
    P = presekPravih(prava(p1, p2), prava(p5, p6))

    # beskonacno daleka tacka po Y
    Q = presekPravih(prava(p3, p7), prava(p2, p6))

    p = prava(p3, P)

    q = prava(p8, Q)

    p4 = presekPravih(p, q)

    return p4
```
